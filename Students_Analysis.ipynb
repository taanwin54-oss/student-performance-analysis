### Study Time vs Performance

We analyze whether students who study more tend to score better grades.

```python
df[["studytime","G3"]].corr()
```

### Gender Performance Comparison

We compare the average grades of male and female students.

```python
df.groupby("sex")["G3"].mean()
```

Visualizations
### Distribution of Final Grades

A histogram helps us understand how grades are distributed among students.

```python
plt.hist(df["G3"],bins=10)
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grades")
plt.ylabel("no of students")
plt.show()
```

Scatter Plot (Study Time vs Grades)

```python
sns.scatterplot(x="studytime",y="G3",data=df)
plt.title("Study Time vs finalgrade")
plt.xlabel("Studytime")
plt.ylabel("finalgrade")
plt.show()
```

Bar Chart (Gender vs Average Score)

```python
df.groupby("sex")["G3"].mean().plot(kind="bar")
plt.title("Gender vs Average Score")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()
```
